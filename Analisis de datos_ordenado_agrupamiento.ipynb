```python
## Seleccion de Columnas relevantes
```

```python
col_analisis = ['Carrera', 'Salario Medio Principiante','Salario Medio con Experiencia']
```

```python
data_analisis = data[col_analisis]
data_analisis.head()
```

```python
## Estadisticas Basicas
```

```python
## Promedio
prom_principantes = data_analisis['Salario Medio Principiante'].mean()
```

```python
## Mediana
media_principantes = data_analisis['Salario Medio Principiante'].median()
```

```python
## Promedio
prom_experienc = data_analisis['Salario Medio con Experiencia'].mean()
media_experienc = data_analisis['Salario Medio con Experiencia'].median()
```

```python
prom_principantes,media_principantes,prom_experienc,media_experienc
```
