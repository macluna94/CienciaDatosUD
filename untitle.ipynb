```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Simulaci√≥n simple
np.random.seed(0)
area = np.random.normal(100, 20, 100)
habitaciones = area / 20 + np.random.normal(0, 0.5, 100)
precio = 50000 + (area * 300) + np.random.normal(0, 10000, 100)
```

```python
df = pd.DataFrame({
    "area": area,
    "habitaciones": habitaciones,
    "precio": precio
})
```

```python
df.head()
```

```python
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlaci√≥n entre variables")
plt.show()
```
