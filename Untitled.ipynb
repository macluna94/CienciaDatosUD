```python
# Importacion de librerias
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Carga del dataframe
df = pd.read_csv("C:/Users/manuell/Videos/CienciaDatos_AN/peliculas.csv")
```

```python
def toNum(val):
    nval = val.replace('$','').replace(',','')
    return pd.to_numeric(nval, errors='coerce')
```

```python
# Convertir datos
# String a Fecha
df['Estreno'] = pd.to_datetime(df['Estreno'])
# String a entero
df['Presupuesto'] = df['Presupuesto'].apply(toNum)
df['TaquillaInternacional'] = df['TaquillaInternacional'].apply(toNum)
df['TaquillaEEUU'] = df['TaquillaEEUU'].apply(toNum)
```

```python
df.info()
```

```python
ndf = df[df['Estreno'] <= '01-05-2018']
```

```python
ndf.sort_values('Estreno', ascending=False)
```

```python
TaquillaTotal = ndf['TaquillaInternacional'] + ndf['TaquillaEEUU']
```

```python
ndf.insert(loc = 1, column = 'TaquillaTotal', value = TaquillaTotal.tolist())
```

```python
y = ndf[['TaquillaTotal']]
```

```python
x = ndf[['Presupuesto']]
```

```python
modelo = LinearRegression()
```
