```python
import pandas as pd 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("C:/Users/manuell/Videos/CienciaDatos_AN/datosmexico.csv")
```

```python
# Definicion de Variable Independiente y Dependiente
```

```python
x = df[["FuerzaLaboral"]]
```

```python
y = df[["PIBperCapita"]]
```

```python
# Creacion de la instancia del modelo
```

```python
modelo = LinearRegression()
```

```python
# Ajuste del modelo
```

```python
modelo.fit(x,y)
```

```python
interceptor = modelo.intercept_
```

```python
coeficiente = modelo.coef_[0]
```
