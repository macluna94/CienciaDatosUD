```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/manuell/Videos/CienciaDatos_AN/peliculas.csv')
```

```python
df.info()
```

```python
def convnum(val):
    nval = val.replace('$','').replace(',','')
    return pd.to_numeric(nval, errors='coerce')
```

#### Conversion de datos

```python
df['Presupuesto'] = df['Presupuesto'].apply(convnum)
df['TaquillaInternacional'] = df['TaquillaInternacional'].apply(convnum)
df['TaquillaEEUU'] = df['TaquillaEEUU'].apply(convnum)
```

```python
df['Estreno'] = pd.to_datetime(df['Estreno'])
```
