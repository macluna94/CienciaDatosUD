```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,50,20)
y = np.sqrt(x)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, color='red')
plt.title("G1")
plt.subplot(1,2,2)
plt.plot(y,x, color='blue')
plt.title("G2")
plt.show()
```

```python
## Orientado a Obejtos
```

```python
grafico = plt.figure()
ejes = grafico.add_axes([0,0,1,1])
ejes.plot(x,y)
ejes.set_xlabel('Eje x')
ejes.set_title('Grafico Objeto')
plt.show()
```

```python
## Superposision de Graficos
```

```python
grafico2 = plt.figure()
ejes1 = grafico2.add_axes([0,0,1,1])
ejes1.plot(x,y)
ejes2.set_title("G1")
ejes2 = grafico2.add_axes([0.5,0.2,0.5,0.3])
ejes2.plot(y,x, color ='red')
ejes2.set_title("G2")
plt.show()
```

```python
## SubPlots con POO
```

```python
graficos, ejes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
graficos, ejes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
```

```python
graficos, ejes = plt.subplots(nrows=1,ncols=3)

for este_eje in ejes:
    este_eje.plot(x,y)
```

```python
graficos, ejes = plt.subplots(nrows=1,ncols=3)
ejes[0].plot(x,y)
ejes[0].set_title("G1")
ejes[2].plot(x,y)
ejes[2].set_title("G3")
plt.show()
```

```python
graf = plt.figure(figsize=(9,3), dpi=50)
ejes = graf.add_axes([0,0,1,1])
ejes.plot(x,y)
```

```python
grafics, ejs = plt.subplots(nrows=2, ncols=1, figsize=(9,4), dpi=70)
ejs[0].plot(x,y)
ejs[1].plot(y,x)
plt.tight_layout()
```

```python
## GRAFICOS CON VIETAS
```

```python
grafico = plt.figure()
ejes = grafico.add_axes([0,0,1,1])
ejes.plot(x, 2*x, label = '2x', linestyle = 'dotted')
ejes.plot(x, x+2, label = 'x+2')
ejes.plot(x, y*3, label = 'Y', linestyle = '--')
ejes.legend()
```

```python
grafico.savefig('migrafico.jpg', dpi=600, )
```

```python
## MArcadores
```

```python
grafico = plt.figure()
ejes = grafico.add_axes([0,0,1,1])
ejes.plot(x, y*3, label = 'Y', linestyle = '--', marker='o', markerfacecolor = 'red')
ejes.legend()
```

```python
grafico = plt.figure()
ejes = grafico.add_axes([0,0,1,1])
ejes.plot(x, y*3, label = 'Y', linestyle = '-', marker='x', markerfacecolor = 'red')
ejes.set_xlim([0,15])
ejes.set_ylim([0,10])
ejes.legend()
```
