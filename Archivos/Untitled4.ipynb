```python
#Funcion para convertir los valores a tipo entero
def covNum(val):
    nval = val.replace('$','').replace(',','')
    return pd.to_numeric(nval, errors = 'coerce')
```

```python
# Pasando los datos string a tipo int
df['Presupuesto'] = df['Presupuesto'].apply(covNum)
df['TaquillaInternacional'] = df['TaquillaInternacional'].apply(covNum)
df['TaquillaEEUU'] = df['TaquillaEEUU'].apply(covNum)
```

```python
df['Estreno'] = pd.to_datetime(df['Estreno'])
```
